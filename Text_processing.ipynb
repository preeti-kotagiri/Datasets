```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"/content/drive/MyDrive/2025/CMR/spam.csv",
                   encoding="iso-8859-1")
```

```python
data
```

```python
data.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1,inplace=True)
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(data.v2,data.v1,
                                                 test_size=0.2,
                                                 random_state=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
X_train_transform = tfidf.fit_transform(X_train)
```

```python
X_train_transform
```

```python
X_train.shape
```

```python
tfidf.get_feature_names_out()
```

```python
X_train_transformed = pd.DataFrame(X_train_transform.toarray(),
              columns=tfidf.get_feature_names_out())
```

```python
X_test_transform = tfidf.transform(X_test)
```

```python
X_test_transform
```

```python
X_test_transformed = pd.DataFrame(X_test_transform.toarray(),
              columns=tfidf.get_feature_names_out())
```

### Building model

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
bnb = BernoulliNB()
```

```python
bnb.fit(X_train_transformed,y_train)
```

```python
y_pred = bnb.predict(X_test_transformed)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
